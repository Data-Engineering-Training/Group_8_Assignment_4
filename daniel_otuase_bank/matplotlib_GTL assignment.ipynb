### 2. Pie plot activity

```python
Workers=[40,35,100,70]
Education=['Primary','JHS','SHS','Tertiary']
```

### 3. Scatter plot activity
Create a scatter plot with matplotlib that displays xt paired with yt and zt:

```python
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt
```

Change the shape of the markers to
1. square for yt
2. triangle for zt
<br> Also:
- Add a label to indicate which points are yt and zt.
- add title
- xlabel and ylabel

```python
plt.scatter(x=xt, y=yt, marker='x')
plt.scatter(x=xt, y=zt, marker='>')
plt.legend('yz')
plt.title('Scatter plot of xt and zt');
plt.xlabel('x values');
plt.ylabel('y values');
```
